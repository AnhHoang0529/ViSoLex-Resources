```python
import pandas as pd
import numpy as np
import json
import os
```

```python
path = "/content/drive/MyDrive/LexNorm/KLTN/Data/ViLexNorm/dictionary/sorted_dict.json"
```

```python
with open(path, 'r') as f:
  data = json.load(f)
```

```python
# Calculate the total number of keys
total_keys = len(data)
print("Từ điển có {} từ.".format(total_keys))
```

```python
# Initialize counters for the number of keys with values containing different token counts
one_value_count = 0
two_value_count = 0
three_value_count = 0
more_value_count = 0

# Count the number of keys with values containing 1, 2, or more value
for value in data.values():
    value_count = len(value)
    if value_count == 1:
        one_value_count += 1
    elif value_count == 2:
        two_value_count += 1
    elif value_count == 3:
        three_value_count += 1
    else:
        more_value_count += 1

print("Từ điển có {} từ có 1 giá trị.".format(one_value_count))
print("Từ điển có {} từ có 2 giá trị.".format(two_value_count))
print("Từ điển có {} từ có 3 giá trị.".format(three_value_count))
print("Từ điển có {} từ có nhiều hơn 3 giá trị.".format(more_value_count))
```

```python
# Initialize counters for the number of keys with values containing different token counts (1-1, 1-n, n-1, n-n)
one_to_one_count = 0
one_to_n_count = 0
n_to_one_count = 0
n_to_n_count = 0

# Count the number of keys with values containing 1-1, 1-n, n-1, or n-n
for key, value in data.items():
    key_tokens = len(key.split())
    value_count = len(value)
    for token in value:
        token_count = len(token.split())
        if key_tokens == 1 and token_count == 1:
            one_to_one_count +=1
        elif key_tokens == 1 and token_count > 1:
            one_to_n_count += 1
        elif key_tokens > 1 and token_count == 1:
            n_to_one_count += 1
        elif key_tokens > 1 and token_count > 1:
            n_to_n_count += 1

print("Từ điển có {} từ có 1-1 giá trị.".format(one_to_one_count))
print("Từ điển có {} từ có 1-n giá trị.".format(one_to_n_count))
print("Từ điển có {} từ có n-1 giá trị.".format(n_to_one_count))
print("Từ điển có {} từ có n-n giá trị.".format(n_to_n_count))
```
